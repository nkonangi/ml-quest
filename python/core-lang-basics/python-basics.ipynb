## BMI Calculator
### The body mass index (BMI) is a measure used in medicine to see if someone is underweight or overweight. This is the formula used to calculate it:
### bmi is equal to the person's weight divided by the person's height squared.



```python
weight = 84
height = 1.65 
bmi = weight/height ** 2
print(bmi)
```

```python
print(round(bmi))
print(round(bmi, 2))
print(round(bmi, 3))
```

## Calculate Tip and sharing bill

```python
bill_amount = float(input("please enter bill amount: "))
tip_percentage = float(input("please enter tip percentage: "))
no_of_people = int(input("How many of you are sharing: "))

total_bill_amount = bill_amount * (1+ tip_percentage/100)

sharing_price = round(total_bill_amount/no_of_people , 2)
print(f"each person share is {sharing_price}")
```

### Variables
#### A Python variable is a reserved memory location to store values.
#### A variable is created the momentyou first assign a value to it.

```python
pi_val = 3.14
pi_val
```

```python
'''
id() function returns the identity of the object.
The identity of an object - Is an integer
    - Guaranteed to be unique
    - Constant for this object during its lifetime.
'''
id(pi_val)
```

```python
# finding Memory address of the variable using hex() and id() functions.
hex(id(pi_val))
```

```python
# let's see same value with multiple variables and their locations
val1 = 50
val2 = 50
temp = val2
val1,type(val1), hex(id(val1))
```

```python
val2,type(val2), hex(id(val2))
```

```python
temp,type(temp), hex(id(temp))
```
